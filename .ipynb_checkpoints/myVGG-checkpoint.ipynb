## VGG16卷积核可视化

### 导入工具库

```python
# 忽略烦人的红色提示
import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei']  # 用来正常显示中文标签  
plt.rcParams['axes.unicode_minus']=False  # 用来正常显示负号
```

### 导入VGG16模型

```python
modelPath='./model.h5'
model = tf.keras.models.load_model(modelPath)
```

### 获取模型参数

```python
model.summary()
```

### 获取模型权重

```python
model.get_weights()
```

```python
len(model.get_weights())
```

```python
for i in range(32):  # 权重 + 偏置项
    print(i, model.get_weights()[i].shape)
```

#### 获取索引为0的卷积核权重

```python
plt.imshow(model.get_weights()[0][:,:,:,0])
```

### 可视化block1_conv1的前64个卷积核

```python
n = 8
fig, axes = plt.subplots(nrows=n, ncols=n, sharex=True,figsize=(15,15))
for i in range(n*n):
    axes[i//n, i%n].imshow(model.get_weights()[0][:,:,:,i])
#     axes[i//n, i%n].set_xlabel('x')
#     axes[i//n, i%n].set_ylabel('y')
    axes[i//n, i%n].set_title('{}'.format(i),fontsize=15)
#     axes[i//8, i%8].xticks([]) # 关闭刻度线显示
#     axes[i//8, i%8].yticks([])
    axes[i//n, i%n].axis('off') # 关闭坐标轴显示
# fig.suptitle(layer_name,fontsize=30)
```

### 可视化block5_conv2的前64个卷积核

```python
n = 8
fig, axes = plt.subplots(nrows=n, ncols=n, sharex=True,figsize=(15,15))
for i in range(n*n):
    axes[i//n, i%n].imshow(model.get_weights()[22][:,:,0,i])
#     axes[i//n, i%n].set_xlabel('x')
#     axes[i//n, i%n].set_ylabel('y')
    axes[i//n, i%n].set_title('{}'.format(i),fontsize=15)
#     axes[i//8, i%8].xticks([]) # 关闭刻度线显示
#     axes[i//8, i%8].yticks([])
    axes[i//n, i%n].axis('off') # 关闭坐标轴显示
# fig.suptitle(layer_name,fontsize=30)
```
